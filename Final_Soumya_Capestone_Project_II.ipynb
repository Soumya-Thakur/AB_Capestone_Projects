## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# write your code to make your dataset analysis ready

class AmazonPrime:
  #Init function
  def __init__(self,df):
    #init method
    try:
      self.df = df.copy()
      self.df['duration_value'] = self.df['duration'].apply(self.set_duration_value)
      self.df['duration unit'] = self.df['duration'].apply(self.set_duration_unit)
      self.cast__ = self.get_cast_or_listed_in('cast')
      self.genres = self.get_cast_or_listed_in('listed_in')
      self.directors = self.getTopValues('director')
      self.countries = self.getTopValues('country')
      self.release_years = self.getTopValues('release_year')
      self.ratings = self.getTopValues('rating')
      self.df['duration_value'] = self.df['duration_value'].apply(lambda x:int(x))
      self.df['duration unit'] = self.df['duration unit'].apply(lambda x:str(x))

    except:
      print('Invalid input!')

  def get_df(self):
    return self.df

  def get_movie_df(self):
    return self.df[self.df['type'] == 'Movie']

  def getTvSows(self):
    return self.df[self.df['type'] == 'TV Show']

  def getTopValues(self,col):
    try:
      return self.df[col].value_counts().sort_values(ascending=False).head(10)
    except:
      return 'You have entered invalid column!'

  def set_duration_value(self, string):
    if string[-7:] == 'Seasons':
      return string[:-8]
    elif string[-6:] == 'Season':
      return string[:-7]
    else:
      return string[:-4]

  def set_duration_unit(self, duration):
    if (duration[-7:] == 'Seasons') | (duration[-6:] == 'Season'):
      return 'Season'
    else:
      return 'min'

  def get_cast_or_listed_in(self,col):
    df = self.df.copy()
    def __get__list():
      list_1=[]
      for i in df[df[col].isna() == False][col]:
        list_1.append(i.split(', '))
      return list_1

    def __get_list_vars(list_1):
      list_2=[]
      for i in list_1:
        for j in i:
          list_2.append(j)
      return list_2

    def __get_dict_vars(list_2):
      unique_var = set(list_2)
      dict_of_vars = dict()
      for i in unique_var:
        dict_of_vars[i] = list_2.count(i)
      return dict_of_vars

    list_1 = __get__list()
    list_2 = __get_list_vars(list_1)
    return pd.Series(__get_dict_vars(list_2)).sort_values(ascending = False)

  def convertInt(self, col):
    return int(col)

  def grtShowsperMonth(self, typ='all'):
    df = pd.DataFrame()
    if typ == 'tv':
      df = self.getTvSows()
    elif typ == 'movie':
      df = self.get_movie_df()
    else:
      df = self.df.copy()

    # Add this line to print the columns and check for 'date_added'
    print("Columns in DataFrame:", df.columns)

    df = df[df['date_added'].isna() == False]
    df['added_month'] = df['date_added'].apply(lambda x:x.split()[0])
    monthly_shows = df['added_month'].value_counts()
    return monthly_shows

  def getTextLenth(self, text):
    return len(text)

  def getDfWithLength(self, col):
    df = self.df.copy()
    df['length'] = df[col].apply(self.getTextLenth)
    return df
````

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Select columns excluding the specified ones
cols_to_exclude = ['imdb_votes', 'tmdb_score', 'imdb_score', 'tmdb_popularity', 'release_year']
cols = [col for col in dataset.select_dtypes(include='number').columns if col not in cols_to_exclude]

# Create pairplot for the remaining numeric columns
sns.pairplot(dataset[cols].dropna())
plt.suptitle('Pair Plot of Selected Numerical Features', y=1.02)
plt.show()
````

I picked the pair plot because it’s a powerful exploratory tool that allows you to simultaneously visualize the relationships and distributions between multiple numerical variables. Unlike single scatterplots or histograms, pair plots show a grid of plots that reveal correlations, trends, clusters, and potential outliers all at once, making it easier to detect patterns that might influence further analysis or modeling.

##### 2. What is/are the insight(s) found from the chart?

From the pair plot, you can gain insights such as which variables have strong linear or non-linear relationships, whether the data is normally distributed or skewed, and if there are distinct groupings or anomalies. For example, seeing tightly clustered points along a diagonal scatterplot indicates strong correlation, while scattered or no clear pattern suggests weak or no correlation.
